```python
def load_df(file):
    """Loads the Dataset"""
    df = pd.read_csv(file)
    return df

def find_categorical_columns(df):
    """Returns name of Columns with string values"""
    categorical_columns = []
    for col in df.columns:
        if type(df[col][0]) == str:
            categorical_columns.append(col)
    return categorical_columns

def encode_columns(df):
    """Apply Encoding to columns"""
    columns = find_categorical_columns(df)
    encoder = LabelEncoder()
    df1 = df.copy()
    for column in columns:
        df1[column] = encoder.fit_transform(df1[column])
    return df1
```

```python
path = 'diabetes_data_upload.csv'
df = load_df(path)
df
```

```python
df = encode_columns(df)
df
```

```python
df['Age'] = df['Age'] /df['Age'].abs().max()
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1:]
```
